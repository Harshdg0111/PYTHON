```python
import pandas as pd 
import numpy as np
```

```python
nvidia = pd.read_csv(r"c:\Users\hdg01\Downloads\nvda_stock_data.csv", index_col=None)
```

```python
nvidia.head()
```

```python
dfs = [nvidia]
```

```python
for df in dfs:
    df['MA50'] = df['Close'].rolling(50).mean()
    df['MA200'] = df['Close'].rolling(200).mean()
```

```python
nvidia.head(200)
```

```python
for df in dfs :
  df['Previous_Day_Close_Price'] = df['Close'].shift(1)
```

```python
nvidia.head()
```

```python
for df in dfs :
  df['Change_In_Price'] = df['Close'] - df['Previous_Day_Close_Price']
```

```python
nvidia.head()
```

```python
for df in dfs :
  df['Percent_Change_In_Price'] = df['Close'].pct_change()
```

```python
nvidia.head()
```

```python
for df in dfs :
  df['Previous_Day_Volumn'] = df['Volume'].shift(1)
```

```python
nvidia.head()
```

```python
for df in dfs : 
  df['Change_In_Volume'] = df['Volume'] - df['Previous_Day_Volumn']
```

```python
nvidia.head()
```

```python
for df in dfs:
  df['Percent_Change_In_Volume'] = df['Volume'].pct_change()
```
